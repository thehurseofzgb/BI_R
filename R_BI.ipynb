# 主成成分分析

```r

data = read.table("2013年-2016年体测单项得分（抽样）.csv",header = TRUE,sep = ",",stringsAsFactors = FALSE)
data_b <- data[data$性别 == "男",c(5,6,7,8,9,10,11)]

```

```r
fa.parallel(cor(data_b),fa = "pc" , n.iter = 100,show.legend = FALSE,main = "碎石图")
```

```r
pc <- principal(cor(data_b),nfactors = 4,rotate = "none")
pc
```

```r
rc <- principal(cor(data_b),nfactors = 4,rotate = "varimax")
rc
```

# 获取组成成分

```r
rc <- principal(cor(data_b),nfactors = 4,rotate = "varimax")
round(unclass(rc$weights),2)
```
